```python
import h5py
import matplotlib.pyplot as plt
```

```python
f= h5py.File("kappa-m111111.hdf5")
list(f)
```

```python
f['temperature'][:]
```

```python
f['kappa'][:,0]
```

```python
temp = f['temperature'][:]
kappa = f['kappa'][:,0]
plt.plot(temp,kappa)
plt.xlim(300,1000)
plt.ylim(0,4)
plt.xlabel("temperature (K)")
plt.ylabel("kappa (W/K m)")
plt.grid()
plt.savefig("0xx-l-thermalcond.jpg")
```

```python
temp = f['temperature'][:]
kappa = f['kappa'][:,1]
plt.plot(temp,kappa)
plt.xlim(300,1000)
plt.ylim(0,4)
plt.xlabel("temperature (K)")
plt.ylabel("kappa (W/K m)")
plt.grid()
plt.savefig("1yy-l-thermalcond.jpg")
```

```python
temp = f['temperature'][:]
kappa = f['kappa'][:,2]
plt.plot(temp,kappa)
plt.xlim(300,1000)
plt.ylim(0,1)
plt.xlabel("temperature (K)")
plt.ylabel("kappa (W/K m)")
plt.grid()
plt.savefig("2zz-l-thermalcond.jpg")
```

```python
temp = f['temperature'][:]
kappa_0 = f['kappa'][:,0]
kappa_1 = f['kappa'][:,1]
kappa_2 = f['kappa'][:,2]
p1 = plt.plot(temp,kappa_0, label='kappa_xx')
p2 = plt.plot(temp,kappa_1, label='kappa_yy')
p3 = plt.plot(temp,kappa_2, label='kappa_zz')
l1 = plt.legend(loc='upper right')
plt.gca().add_artist(l1)
plt.xlim(300,1000)
plt.ylim(0,4)
plt.xlabel("temperature (K)")
plt.ylabel("kappa (W/K m)")
plt.title("lattice thermal conductivity")
plt.grid()
plt.savefig("l-thermalcond.jpg", dpi=600)
```
